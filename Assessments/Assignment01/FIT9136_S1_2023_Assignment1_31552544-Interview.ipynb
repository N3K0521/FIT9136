# Task 2. Create Deck function

The purpose of this function is to generate a deck of cards based on the provided arguments, which include the deck, suits, and values. This function does not have a return value.

```python
# According to the requirements, the arguments take by this function is: deck, suits, values
def create_deck(deck, suits, values):
    """
    This function creates a new deck of cards. The deck is a list
    that stores all the cards generated from a complete combincation
    of suits and values. Suits are the type of cards, while values are
    just simply values of the cards. 
    
    Parameters
    ----------
    Deck: a list
    Suits: a list of card suit symbols
    Values: a fixed list of card values includes 2-10 and J, Q, K, A
    
    Returns
    -------
    None
    """
    # nested loop to print out each card
    for suit in suits:
        for value in values:
            # The format used: value of suit. e.g. 2 of ‚ô•
            card = f"{value} of {suit}"
            deck.append(card)
    print(deck) 
```

# Task 3. Shuffle Deck function

This function takes two arguments: deck, which is a list representing a deck of cards, and
suits, which is a list of the different suits used in the deck. The function shuffles the cards in
the deck randomly and displays the shuffled deck to the user. After shuffling, the deck
always maintains the order where the first card is the 'A' card of the first suit in the suits list,
the middle position (i.e., round((n+1)/2, 0))) is the 'Q' card of the second suit in the suits list, and

2
the last card is the 'K' card of the last suit in the suits list. If a required card ('A', 'Q', or 'K' of
any suit) is picked by the player, that card is ignored. For example, if suits = ["‚ô•", "‚ô¶", "‚ô£",
"‚ô†"], after shuffling the deck might be ["A of ‚ô•", ‚Ä¶‚Ä¶, "Q of ‚ô¶"......, "K of ‚ô†"]. This function
does not return anything.

```python
def shuffle_deck(deck, suits):
    """
    This function shuffles the cards in the deck randomly and dislays
    the shuffled deck to the user. The deck always maintains the order
    where 'A' card is the first card of the first suit in the suits list.
    The 'Q' card is the second suit in the suits list.
    The last card is the 'K' card.
    This function does not return anything.
    
    Parameters
    ----------
    Deck: a list
    Suits: a list of card suit symbols
    
    Returns
    -------
    None
    """
    
    # This method Takes a list and reorganise the order of the items.
    # It changes the original list without returning a new list
    random.shuffle(deck)

    # Ensuring 'A', 'Q', or 'K' of any suit are in the required orders
    # In this case, the index() method can be used to search the element
    a_index = deck.index(f"A of {suits[1]}") # 'A' is always in the 1st 
    ten_index = deck.index(f"10 of {suits[0]}") # 'Q' is in the second suit
    j_index = deck.index(f"J of {suits[-1]}") # 'K' is always in the last

    deck[j_index], deck[0] = deck[0], deck[j_index]
    deck[ten_index], deck[round((len(deck) + 1) / 2) - 1] = deck[round((len(deck) + 1) / 2) - 1], deck[ten_index]
    deck[a_index], deck[-1] = deck[-1], deck[a_index]
    
    # User-friendly interface
    print("The deck has been successfully shuffled. \n")
```

# Task 4. Pick Card function

This function takes one argument, deck. It randomly selects one card from the deck and returns it. The picked card is then removed from the deck. Both the player and the robot will be picking cards from the same deck.

```python
# The choice() method returns a randomly selected element from the specified sequence.
# The argument takes by this function is deck.
def pick_card(deck):
    """
    Randomly select one card from the deck and return it.
    Then remove the chosen card from the deck.
    
    Parameters
    ----------
    Deck: a list
    
    Returns
    -------
    card: an element of the deck list
    """
    card = random.choice(deck)
    deck.remove(card)
    print(card)
    
```

# Task 6. Check Result function


This function accepts three parameters - player_cards, robot_cards, and suits. Its main task is to compare the combination of cards held by the player and the robot to determine if the player has won the game. The robot is considered as another player. To win, the player must meet one of the rule in the following order of priority: 
Rule 1 > Rule 2 > Rule 3 > Rule 4.

```python
def check_result(player_cards, robot_cards, suits):
    """
    This function compare the combination of cards held by the
    player and the robot to determine if the player has won
    the game. To win the player must meet one of the 4 rules in
    the order of priority: Rule 1 > Rule 2 > Rule 3 > Rule 4
    
    The rules are listing as below:
    Rule 1: Holding the same value card for all the defined suits
    Rule 2: Same values for at least (total defined suits - 1)
    Rule 3: Holding more cards from second_suit (suits[1]) than the robot
    Rule 4: Holding a higher average of cards value than robot
    
    Parameters
    ----------
    player_cards: list
    robot_cards: list
    suits: list
    
    Returns
    -------
    A boolean value - True or False -> player win/lose
    
    Examples:
    Meet rule 1:
    Suits: ["‚ô•", "‚ô¶", "‚ô£", "‚ô†"]
    Player's cards: : ['A of ‚ô•', 'A of ‚ô¶', 'A of ‚ô£', 'A of ‚ô†']
    >>>True
    
    Didn't meet rule 1
    Meet rule 2:
    Suits: ["‚ô•", "‚ô¶", "‚ô£", "‚ô†"]
    Player's cards: : ['A of ‚ô•', 'A of ‚ô¶', 'A of ‚ô£']
    >>>True
    
    Didn't meet rule 2
    Meet rule 3:
    Suits: ["‚ô•", "‚ô¶", "‚ô£", "‚ô†"]
    Player's cards: ['A of ‚ô•', '2 of ‚ô¶', '3 of ‚ô£', '4 of ‚ô¶']
    Robot cards: ['2 of ‚ô•', '3 of ‚ô¶', '4 of ‚ô£', '5 of ‚ô†']
    >>>True
    
    Didn't meet rule 3
    Meet rule 4:
    Suits: ["‚ô•", "‚ô¶", "‚ô£", "‚ô†"]
    Player's cards: ['10 of ‚ô•', '2 of ‚ô¶', '3 of ‚ô£', '4 of ‚ô¶']
    Robot cards: ['2 of ‚ô•', '3 of ‚ô¶', '4 of ‚ô£', '5 of ‚ô†']
    >>>True
    
    """
    def count_values(cards):
        counts = {} #dictionary
        for card in cards:
            value = card.split()[0]
            suit = card.split()[2]
            if value not in counts:
                counts[value] = 0
            counts[value] += 1
            if suit not in counts:
                counts[suit] = 0
            counts[suit] += 1
        return counts

    def average_value(cards):
        total = 0
        for card in cards:
            value = card.split()[0]
            # converting the card value to integers
            if value == "A":
                total += 1
            elif value == "J":
                total += 11
            elif value == "Q":
                total += 12
            elif value == "K":
                total += 13
            else:
                total += int(value)
        return total / len(cards)

    player_counts = count_values(player_cards)
    robot_counts = count_values(robot_cards)

    # Rule 1
    # The player holds the same value card for all the defined suits.
    for value, count in player_counts.items(): # player meets rule1
        if count == len(suits):
            return True
    for value, count in robot_counts.items(): # robot meets rule1
        if count == len(suits):
            return False

    # Rule 2
    # The player has the same values for at least the total defined suits minus one.
    for value, count in player_counts.items():
        if count >= len(suits) - 1:
            return True
    for value, count in robot_counts.items():
        if count >= len(suits) - 1:
            return False

    # Rule 3
    #  The player holds more cards from the suit in position 2 than the robot.
    second_suit = suits[1]
    # get() method returns the value of the item with the specified key
    # syntax: dictionary.get(keyname, value)
    if player_counts.get(second_suit, 0) > robot_counts.get(second_suit, 0):
        return True
    elif robot_counts.get(second_suit, 0) > player_counts.get(second_suit, 0):
        return False

    # Rule 4
    #  The player holds a higher average of the card‚Äôs value than robot.
    if average_value(player_cards) > average_value(robot_cards):
        return True
    elif average_value(robot_cards) > average_value(player_cards):
        return False

    # The player loses if cards don't meet any of the rule
    return False
```
